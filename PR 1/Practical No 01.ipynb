Making Impute function for filling Null values

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

```python
df['Age'] = df[['Age','Pclass']].apply(impute_age,axis=1) # Appyling the function 
```

```python
df.drop('Cabin',axis=1,inplace=True) # Droping Cabin Column becasue here lots of null values so it's not possible to fill na values.
```

```python
df.dropna(inplace=True)
```
